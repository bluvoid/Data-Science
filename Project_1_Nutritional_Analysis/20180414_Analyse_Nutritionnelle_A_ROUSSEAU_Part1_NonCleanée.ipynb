## Alimentation équilibrée  & Apports Journaliers Recommandés = AJR (Cf Wikipedia):

<span style="color:#3366ff"> Une bonne alimentation fournit à l'organisme les nutriments essentiels : fluide, 
acides aminés essentiels des protéines1, acides gras, vitamines, minéraux, et suffisamment de calories. 
 
<span style="color:#3366ff"> Elle fournit l'énergie nécessaire sans exposer à la toxicité.

<span style="color:#3366ff"> Un régime équilibré est également important afin de réduire les risques de santé, comme l'obésité, 
les maladies cardiaques, le diabète de type 2, l'hypertension et le cancer2.  

<span style="color:#3366ff"> Les apports journaliers recommandés (AJR) sont des valeurs-repères utilisées en tant que référence pour l'étiquetage des produits alimentaires, pour fournir une indication sur les apports journaliers en micronutriments 
(vitamines et minéraux) nécessaires à un adulte-type moyen.
